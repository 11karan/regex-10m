```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('/Users/tarunchawrani/Desktop/Data/airlines_flights_data.csv')
df
```

```python
plt.figure(figsize=(10, 6))
airline_counts = df['airline'].value_counts()
print(airline_counts)
sns.barplot(x=airline_counts.index, y=airline_counts.values)
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.title('Number of Flights by Airline')
plt.xticks(rotation=45)
plt.show()
```

```python
departure=df['departure_time'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot( x=departure.index, y=departure.values)
plt.xlabel('Departure Time')
plt.ylabel('Number of Flights')
plt.title('Departure vs Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

```python
Arrival=df['arrival_time'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot( x=Arrival.index, y=Arrival.values)     
plt.xlabel('Arrival Time')
plt.ylabel('Number of Flights')
plt.title('Arrival Time Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
Source_city=df['source_city'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=Source_city.index, y=Source_city.values)
plt.xlabel('Source City')
plt.ylabel('Number of Flights')
plt.title('Number of Flights by Source City')
plt.xticks(rotation=45)
plt.show()
```

```python
Destination_city=df['destination_city'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=Destination_city.index, y=Destination_city.values)
plt.xlabel('Destination City')
plt.ylabel('Number of Flights')
plt.title('Number of Flights by Destination City')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Time_of_Journey']=df['departure_time']+'-'+df['arrival_time']
Time_of_Journey_mean=df.groupby('Time_of_Journey')['price'].mean()
print(Time_of_Journey_mean)
plt.figure(figsize=(10,6))
sns.barplot(x=Time_of_Journey_mean.index, y=Time_of_Journey_mean.values)
plt.xlabel('Time_of_Journey')
plt.ylabel('Averge Price')
plt.xticks(rotation=45)
plt.show()
```

```python
df['journey'] = df['source_city'] + ' - ' + df['destination_city']
journey_mean = df.groupby('journey')['price'].mean()
print(journey_mean)
plt.figure(figsize=(10,6))
sns.barplot(x=journey_mean.index, y=journey_mean.values)
plt.xlabel('Journey')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.title('Average Price by Journey')
plt.show()
```

```python
days_left1 = df[df['days_left'] <= 2]
group_5 = days_left1.groupby('days_left')['price'].mean()
print(group_5)
plt.figure(figsize=(10, 6))
sns.barplot(x=group_5.index, y=group_5.values)
plt.xlabel('Days Left')
plt.ylabel('Average Price')
plt.title('Average Price by Days Left') 
plt.xticks(rotation=45)
plt.show()
```

```python
class_1=df.groupby('class')['price'].mean()
plt.figure(figsize=(10,6))
sns.barplot(x=class_1.index, y=class_1.values)
plt.xlabel('class_1')
plt.ylabel('Average Price')
plt.title('')
plt.xticks(rotation=45)
plt.show()
```
